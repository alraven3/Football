```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
link = 'https://www.fotmob.com/teams/9857/overview/bologna'
res = requests.get(link)
soup = BeautifulSoup(res.content, 'lxml')
scripts = soup.find_all('script')
```

- PPDA: passes allowed per defensive action in the opposition half
- OPPDA: opponent passes per defensive action in the opposition
- Deep completions (DC, "deep"): Passes completed within an estimated 20 yards of goal (crosses excluded)
- Opposition Deep completions (ODC, "deep_allowed"): Opponent passes completed within an estimated 20 yards of goal (crosses excluded)
- xGChain: Total xG of every possession the player is involved in
- xGBuildup: Total xG of every possession the player is involved in without key passes and shots

```python
data = scripts[35].string
before, sep, after = data.partition(',"seo"')
```

```python
def check_json_brackets(json_string):
    square_brackets = 0  # For []
    curly_brackets = 0   # For {}

    for char in json_string:
        if char == '[':
            square_brackets += 1
        elif char == ']':
            square_brackets -= 1
        elif char == '{':
            curly_brackets += 1
        elif char == '}':
            curly_brackets -= 1

    return square_brackets, curly_brackets

# Replace with your actual JSON string
json_string = before
square, curly = check_json_brackets(json_string)

print("Mismatch in square brackets: ", square)
print("Mismatch in curly brackets: ", curly)
```
