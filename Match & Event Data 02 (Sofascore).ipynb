**Team Stats**

```python
response_stats = requests.get('https://api.sofascore.com/api/v1/event/11352508/statistics', headers=headers)
stats = response_stats.json()
stats2 = pd.json_normalize(stats['statistics'][0]['groups'])
stats2
```

```python
xG = pd.json_normalize(stats['statistics'][0]['groups'][0]['statisticsItems']).T
```

```python
xG = pd.json_normalize(stats['statistics'][0]['groups'][0]['statisticsItems'])
possession = pd.json_normalize(stats['statistics'][0]['groups'][1]['statisticsItems'])
shots = pd.json_normalize(stats['statistics'][0]['groups'][2]['statisticsItems'])
misc = pd.json_normalize(stats['statistics'][0]['groups'][3]['statisticsItems'])
creation = pd.json_normalize(stats['statistics'][0]['groups'][4]['statisticsItems'])
passes = pd.json_normalize(stats['statistics'][0]['groups'][5]['statisticsItems'])
duels = pd.json_normalize(stats['statistics'][0]['groups'][6]['statisticsItems'])
defending = pd.json_normalize(stats['statistics'][0]['groups'][7]['statisticsItems'])
match_stats = pd.concat([xG, possession, shots, creation,passes,duels,defending,misc])
match_stats.loc[:,['name','home','away']]
```
