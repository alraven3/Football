```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import time

# Initialise the webdriver
firefox_options = Options()
firefox_options.headless = True
driver = webdriver.Firefox(options=firefox_options)
driver.get('https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats')

time.sleep(5)  # ensure the page is fully loaded

# Render the dynamic content to static HTML
html = driver.page_source
# print(html)

# Parse the static HTML
soup1 = BeautifulSoup(html, "html.parser")

# Close the webdriver
driver.close()
```

# Leaderboards for 77 metrics

```python
#here are the leaderboards
#soup1.find_all('div',{'id':'all_leaders'})
```

```python
data_grid_boxes = soup1.find_all('div', {'class': 'data_grid_box'})

all_leaderboards = []
for box in data_grid_boxes[5:83]:  # assuming indices 5 to 82 represent the leaderboards for different metrics
    metric_name = box.find('caption').text.strip()
    leaderboard_data = []
    for row in box.tbody.find_all("tr"):
        columns = row.find_all("td")
        if columns:
            rank = columns[0].text.strip()
            player = columns[1].text.strip()
            value = columns[2].text.strip()
            leaderboard_data.append([rank, player, value, metric_name])
    leaderboard_df = pd.DataFrame(leaderboard_data, columns=["Rank", "Player", "Value", "Metric"])
    all_leaderboards.append(leaderboard_df)
```

```python
unique_metrics = pd.Series([df['Metric'].unique()[0] for df in all_leaderboards]).unique()
```

```python
#use a dictionary comprehension to construct the dataframe with all the metric names and each of their indices
# Create a dataframe with metric names and their corresponding indices
index_mapping = pd.DataFrame({
    'Metric': unique_metrics,
    'Index': range(5,len(unique_metrics)+5)  # Use the length of unique_metrics to determine the range, # Adjust the range to start from 5 and end at len(unique_metrics) + 5
})

# Set the display option to show all rows
pd.set_option('display.max_rows', None)

# Display the dataframe
print(index_mapping)

# Reset the display option to its default value (optional)
pd.reset_option('display.max_rows')
```

# Other Big 5 Pages/Links

```python
# Function to scrape tables from a single link and store them in named DataFrames
def scrape_tables_from_link(link):
    response = requests.get(link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract specific tables and name them
        extracted_dataframes = {}
        
        # Extract nationality table
        nationality_table = soup.find('table', {'id': 'nations'})  # Replace 'table_id' with the actual ID of the table for nationalities
        if nationality_table:
            extracted_dataframes['nationalities'] = pd.read_html(str(nationality_table))[0]
        else:
            pass
        
        # Extract player wages table
        squad_wages_table = soup.find('table', {'id': 'squad_wages'})  # Replace 'table_id_2' with the actual ID of the player wages table
        if squad_wages_table:
            extracted_dataframes['squad_wages'] = pd.read_html(str(squad_wages_table))[0]
        else:
            pass
        
        player_wages_table = soup.find('table', {'id': 'player_wages'})  # Replace 'table_id_2' with the actual ID of the player wages table
        if player_wages_table:
            extracted_dataframes['player_wages'] = pd.read_html(str(player_wages_table))[0]
        else:
            pass
        # Add more table extractions as needed
        
        return extracted_dataframes
    else:
        print(f"Failed to fetch data from link: {link}")
        return None

# Main function to scrape tables from the first series of links
def scrape_first_series():
    # Define the first series of links
    links = [
        'https://fbref.com/en/comps/Big5/nations/Big-5-European-Leagues-Nationalities',
        'https://fbref.com/en/comps/Big5/wages/Big-5-European-Leagues-Wages'
    ]
    all_data_frames = {}
    
    # Scrape tables from each link
    for link in links:
        data_frames = scrape_tables_from_link(link)
        if data_frames:
            all_data_frames.update(data_frames)
    
    return all_data_frames

# Call the main function to scrape tables from the first series of links
first_series_tables = scrape_first_series()

# Now you can access the DataFrames directly using their names
nationalities_df = first_series_tables['nationalities']
player_wages_df = first_series_tables['player_wages']
squad_wages_df = first_series_tables['squad_wages']
```
