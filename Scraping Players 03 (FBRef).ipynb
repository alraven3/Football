# Full Player/Scouting Report

***This is a scraper for extracting the data from FBRef's player scouting reports, using the example of Bruno Fernandes (as of 18 December 2023).***  
*The goal is to scrape the three columns of the full scouting report: Stats (the column with the names of each metric), Per 90 Values, and Percentiles.*

```python
url = 'https://fbref.com/en/players/507c7bdf/scout/365_m1/Bruno-Fernandes-Scouting-Report'
page = requests.get(url)
soup = BeautifulSoup(page.text,'lxml')
```

```python
table = soup.find('table', id="scout_full_AM")
```

About the table id: 
- it's for the full scouting report ("scout_full_XX"), not the summary one, which is "scout_summary_XX"
- the 'XX' is for the position (group): "GK" for goalkeepers, "CB" for centre-backs, "FB" for full-backs, "MF" for midfielders, "AM" for att. midfielders/wingers, "FW" for centre forwards/strikers/etc.

```python
data_rows = table.find('tbody')
```

```python
for x in data_rows.find_all():
   if len(x.get_text(strip=True)) == 0:
      x.extract()
```

```python
per_90_values = data_rows.find_all('td',{'class':"right"})
stats_column = data_rows.find_all('th',{'class':"right poptip endpoint endpoint"})
percentiles = data_rows.find_all('td',{'class':"left endpoint endpoint tooltip"})
```

```python
df = pd.DataFrame({'Statistic':stats_column, "Per 90": per_90_values, "Percentile":percentiles})
```

```python
df
```

```python
stats = pd.DataFrame(stats_column)
values_per90 = pd.DataFrame(per_90_values)
percentile_list = []
for each in percentiles:
    percentile_list.append(each.text.strip())
percentiles = pd.DataFrame(percentile_list)
stats.columns = ['Statistic']
values_per90.columns = ['Per 90']
percentiles.columns = ['Percentile']
scouting_report = pd.concat([stats,values_per90,percentiles],axis=1,join='inner')
scouting_report
```

```python
table2 = soup.find('table', id="scout_full_MF")
data_rows2 = table2.find('tbody')
for x in data_rows2.find_all():
   if len(x.get_text(strip=True)) == 0:
      x.extract()
per_90_values2 = data_rows2.find_all('td',{'class':"right"})
stats_column2 = data_rows2.find_all('th',{'class':"right poptip endpoint endpoint"})
percentiles2 = data_rows2.find_all('td',{'class':"left endpoint endpoint tooltip"})
stats2 = pd.DataFrame(stats_column2)
values_per902 = pd.DataFrame(per_90_values2)
percentile_list2 = []
for each in percentiles2:
    percentile_list2.append(each.text.strip())
percentiles2 = pd.DataFrame(percentile_list2)
stats2.columns = ['Statistic']
values_per902.columns = ['Per 90']
percentiles2.columns = ['Percentile']
scouting_report2 = pd.concat([stats2,values_per902,percentiles2],axis=1,join='inner')
scouting_report2
```
