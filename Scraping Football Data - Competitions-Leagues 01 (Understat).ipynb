- PPDA: passes allowed per defensive action in the opposition half
- OPPDA: opponent passes per defensive action in the opposition
- Deep completions (DC, "deep"): Passes completed within an estimated 20 yards of goal (crosses excluded)
- Opposition Deep completions (ODC, "deep_allowed"): Opponent passes completed within an estimated 20 yards of goal (crosses excluded)
- xGChain: Total xG of every possession the player is involved in
- xGBuildup: Total xG of every possession the player is involved in without key passes and shots

```python
# Get the fixtures data
strings = scripts[2].string 
# Getting rid of unnecessary characters from json data
ind_start = strings.index("('")+2 
ind_end = strings.index("')") 
json_data = strings[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')
data = json.loads(json_data)
df_teams = pd.DataFrame(data.values())
df_teams
```

```python
df_teams = df_teams.explode("history")
h = df_teams.pop("history")
df_teams = pd.concat([df_teams.reset_index(drop=True), pd.DataFrame(h.tolist())], axis=1)
df_teams = df_teams.infer_objects()
df_teams
```

```python
df_teams.columns
```

```python
table = df_teams.groupby(['title']).agg({'wins': 'sum', 'draws': 'sum', 
   'loses': 'sum', 'scored': 'sum', 'missed': 'sum', 'pts': 'sum', 
   'xG': 'sum', 'xGA': 'sum', 'xpts': 'sum', 'npxG': 'sum', 
 'npxGA': 'sum','npxGD': 'sum', 'deep': 'sum', 'deep_allowed': 'sum'}).reset_index()
table.sort_values('pts',ascending=False)
```

```python
import asyncio
import json
import aiohttp
import nest_asyncio
from understat import Understat
# to avoid errors apply the nest asyncio 
nest_asyncio.apply()
```

```python
async with aiohttp.ClientSession() as session:
        understat = Understat(session)
        table = await understat.get_league_table("EPL", "2023")
        df = pd.DataFrame(table)
```

```python
df.columns = df.iloc[0]
df = df[1:]
```
