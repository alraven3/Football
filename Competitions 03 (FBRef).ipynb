```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import time

url = "https://fbref.com/en/comps/882/Europa-Conference-League-Stats"

# Initialise the webdriver
firefox_options = Options()
firefox_options.headless = True
driver = webdriver.Firefox(options=firefox_options)
driver.get(url)

time.sleep(5)  # ensure the page is fully loaded

# Render the dynamic content to static HTML
html = driver.page_source
# print(html)

# Parse the static HTML
soup1 = BeautifulSoup(html, "html.parser")

# Close the webdriver
driver.close()
```

```python
soup1.find_all('div',{'class':'match-summary'}) #these are the current fixtures - here KO playoffs
```

```python
#here are the leaderboards
#soup1.find_all('div', {'class':'data_grid_box'}) 
```

```python
goals = soup1.find_all('div', {'class':'data_grid_box'})[0] #the first one is for goals
goal_leaders = pd.DataFrame(
    columns=["Rank", "Player", "Goals"]
)

mylist = []
for goals_row in goals.tbody.find_all("tr"):
    goals_columns = goals_row.find_all("td")

    if goals_columns != []:
        rank = goals_columns[0].text.strip()
        player = goals_columns[1].text.strip()
        goals_nb = goals_columns[2].text.strip()

        mylist.append([rank, player, goals_nb])

goal_leaders = pd.DataFrame(
    mylist,
    columns=["Rank", "Player", "Goals"]
)
goal_leaders
```

```python
PSxGDF90 = soup1.find_all('div', {'class':'data_grid_box'})[77] #the last one is for PSxG-GA/90
PSxGDF90_leaders = pd.DataFrame(
    columns=["Rank", "Player", "PSxG-GA/90"]
)

mylist = []
for PSxGDF90_row in PSxGDF90.tbody.find_all("tr"):
    PSxGDF90_columns = PSxGDF90_row.find_all("td")

    if PSxGDF90_columns != []:
        rank = PSxGDF90_columns[0].text.strip()
        player = PSxGDF90_columns[1].text.strip()
        psxgdf90 = PSxGDF90_columns[2].text.strip()

        mylist.append([rank, player, psxgdf90])

PSxGDF90_leaders = pd.DataFrame(
    mylist,
    columns=["Rank", "Player", "PSxG-GA/90"]
)
PSxGDF90_leaders
```

```python
nations_soup = soup1.find('table', {'id':'nations'})
```

```python
df_nationalities = pd.DataFrame(
    columns=["Nation", "# Players", "Min","List"]
)

mylist = []
for nations_row in nations_soup.tbody.find_all("tr"):
    nations_columns = nations_row.find_all("td")

    if nations_columns != []:
        nation = nations_columns[0].text.strip()
        players = nations_columns[1].text.strip()
        Min = nations_columns[2].text.strip()
        List = nations_columns[3].text.strip()

        mylist.append([nation, players, Min, List])

df_nationalities = pd.DataFrame(
    mylist,
    columns=["Nation", "# Players", "Min","List"]
)
```
